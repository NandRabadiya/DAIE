```python
import numpy as np
import pandas as pd
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
#initializing the data 
x=[110,120,130,140]
y=[20,25,35,55]


#for line char we will use the plot()
plt.plot(x,y,'-',color='pink')
plt.title("Line Chart")
plt.ylabel('Weight in pound')
plt.xlabel('Height in cm')

# If figure is not hrere then there will be in the same graph both the plot
plt.figure()
a=np.array([40,50,60,76])
b=a*2

plt.plot(a,b,'-.',color='purple')
plt.show()
```

## Histogram

```python
rand_list = []
for i in range(0,10):
    data = np.random.randint(30)
    rand_list.append(data)
    print(data)
```

```python
# plotting a basic histogram

plt.hist(rand_list,bins=4,color='skyblue',edgecolor='red')

```

## Scatter plot

```python
x = np.array([12,23,45,67,34,34])
y = np.array([45,67,35,7,34,68])

b = np.array([12,23,45,67,34,34])
a = np.array([45,67,35,7,34,68])

bubble_size = [66,88,111,142,200,240]



plt.scatter(a,b,color='orange',label="Group 1",s=bubble_size,alpha=0.9,edgecolor='b',linewidths=2)
plt.scatter(x,y,label='Group 2',s=bubble_size,alpha=0.4,edgecolor='b',linewidths=2)

plt.legend()
plt.show()
```

## Pie Chart

```python
cars= ['AUDI','BMW','FORD','TESLA','TATA']
data = [23,34,56,78,99]


# creating the plot
fig = plt.figure(figsize=(5,3))
plt.pie(data,labels=cars)
plt.show()
```

## Bar cHART

```python
fruits = ['Apples', 'Bananas', 'Kiwi', 'Mango']
sales = [400, 350, 300, 450]

bar_labels=['red','blue','green','brown']
bar_color=['tab:red','tab:blue','tab:green','tab:brown']


plt.bar(fruits, sales,label=bar_labels,color=bar_color)
plt.title('Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('Sales')

plt.legend(title = 'Fruit COlor')
plt.show()
```

## Iris DATASET

```python
import sklearn.datasets as data
df=data.load_iris()
arr=np.array(df)
x=df.data[:,0]
y=df.data[:,1]
plt.scatter(x,y)
plt.show()
```

```python

iris=data.load_iris()
_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)
```
